# .loc, um método muito importante

    -Pegar 1 linha
    -Pegar linhas de acorda com alguma condição
    -Pegar linhas e colunas específicas
    -Pegar 1 valor específico

```python
# pegar uma linha
display(tabelaVendas.loc[1])

#correspondendo à uma condição
vendasNorteShopping = tabelaVendas.loc[tabelaVendas['ID Loja'] == 'Norte Shopping']

#pegar várias linhas e colunas usando o loc
vendasNorteShopping = tabelaVendas.loc[tabelaVendas['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendasNorteShopping)

#pegar um valor específico
print(tabelaVendas.loc[1, 'Produto'])
```

# Adicionar 1 coluna

```python
# a partir de uma colina que ja existe
tabelaVendas['Comissão'] = tabelaVendas['Valor Final'] * 0.05
#diplay(tabelaVendas)


# criar uma coluna com valor padrão
tabelaVendas.loc[:,'Imposto'] = 0
display(tabelaVendas)
```

```python
tabelaVendasDez = pd.read_excel('Vendas - Dez.xlsx')
tabelaVendasDez['Comissão'] = tabelaVendasDez['Valor Final'] * 0.05
tabelaVendasDez.loc[:, 'Imposto'] = 0
#display(tabelaVendasDez)

tabelaVendas = tabelaVendas.append(tabelaVendasDez)
display(tabelaVendas)
```

# Excluindo Linhas e colunas

```python
tabelaVendas = tabelaVendas.drop(0, axis = 0)
display(tabelaVendas)
```
